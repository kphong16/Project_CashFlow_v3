```python
#### INPUT BASIC DATA ####
mtrt = 26 # months
prd_prjt = 24 # months

prcs_rate = pd.Series(np.ones(prd_prjt) * 1/prd_prjt, index=np.arange(1, prd_prjt+1))
#### INPUT BASIC DATA ####


#### INPUT COST DATA ####
costA = acc.account(title="costA", tag=("costA", "-"), mtrt=mtrt,
                    add_scdd_ipt=pd.Series([30_000.000], index=[0]),
                    note="-")

costB_amt = 40_000.000
costB = acc.account(title="costB", tag=("costB", "-"), mtrt=mtrt,
                    add_scdd_ipt=costB_amt * prcs_rate,
                    note="-")

dct = {'costA':costA, 'costB':costB}
cost = acc.merge(dct)
#### INPUT COST DATA ####
```
